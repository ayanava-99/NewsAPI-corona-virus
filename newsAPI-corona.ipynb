```python
import requests
import time
import datetime


url = ('http://newsapi.org/v2/top-headlines?'#fetching the top headlines 
        'language=en&' #specify the language code 
        'q=corona&'    #specify the query
        'apiKey=ENTER THE API KEY')

```

```python
while True: 
    start_time = time.time()
    now = datetime.datetime.now()
    response = requests.get(url)
    data = response.json()
    articles = data["articles"] 
    titles = [] 
    urls=[]
    des=[]
    
    
    print('====================')
    print('Updated Last:',now.strftime("%H:%M:%S") )
    print('====================')
    print("Time consumed %s seconds" % (time.time() - start_time))
    print("============================================================================================================")
    
    for ar in articles: 
        titles.append(ar["title"]) 
        urls.append(ar['url'])
        des.append(ar['description'])
    
    for i in range(len(titles)):
        
        print(i+1,'->',titles[i],'')
        print(des[i])
        print(urls[i],'\n')
    print("============================================================================================================")
    time.sleep(1800)#used to add delay in the execution of the program by 1800 seconds 
```
