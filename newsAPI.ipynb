```python
import requests
import time
import datetime
url = ('http://newsapi.org/v2/top-headlines?'
        'language=en&'
        'q=corona&'
        'apiKey=8767ed3af4c449c9b56c2df1ec9aeac5')

while True:
    start_time = time.time()
    now = datetime.datetime.now()
    response = requests.get(url)
    data = response.json()
    articles = data["articles"] 
    titles = [] 
    urls=[]
    des=[]
    
    
    print('====================')
    print('Updated Last:',now.strftime("%H:%M:%S") )
    print('====================')
    print("Time consumed %s seconds" % (time.time() - start_time))
    print("============================================================================================================")
    
    for ar in articles: 
        titles.append(ar["title"]) 
        urls.append(ar['url'])
        des.append(ar['description'])
    
    for i in range(len(titles)):
        
        print(i+1,'->',titles[i],'')
        print(des[i])
        print(urls[i],'\n')
    print("============================================================================================================")
    time.sleep(15)
```
